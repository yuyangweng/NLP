```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

count=CountVectorizer()
docs=np.array([
        'The sun is shining',
        'The weather is sweet',
        'The sun is shining, the weather is sweet, and one and one is two'])

bag=count.fit_transform(docs)
print(bag.shape)
print()
print(count.vocabulary_)
print()
print(bag.toarray())
# print()
# print(bag)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf=TfidfTransformer(use_idf=True,norm='l2',smooth_idf=True)
bag=count.fit_transform(docs)
print(tfidf.fit_transform(bag).toarray())
print()
print(tfidf.fit_transform(bag))
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf=TfidfTransformer(use_idf=True,norm=None,smooth_idf=True)
bag=count.fit_transform(docs)
print(tfidf.fit_transform(bag).toarray())
print()
print(tfidf.fit_transform(bag))
print(tfidf.fit_transform(bag).toarray().shape[0])
```

```python
l2_tfidf=[]
for row in range(tfidf.fit_transform(bag).toarray().shape[0]):
  row_tfidf=tfidf.fit_transform(bag).toarray()[row]
  l2_tfidf.append(row_tfidf/np.sqrt(np.sum(row_tfidf**2)))
l2_tfidf=np.array(l2_tfidf)
print(l2_tfidf)
```

```python
from nltk.stem.porter import PorterStemmer

porter=PorterStemmer()

def tokenizer(text):
  return text.split()
def tokenizer_porter(text):
  return [porter.stem(word) for word in text.split()]
```

```python
print(tokenizer('runners like running and thus they run'))
print(tokenizer_porter('runners like running and thus they run'))
```
